```python
"""Use FastAPI with ngrok.

origin: Code for the Python Developer Training code challenge.
date: 07/05/2024
author: Marcelo Andrade
"""

# Instal dependencies
!pip install flask
!pip install flask_ngrok pyngrok
!pip install fastapi nest-asyncio pyngrok uvicorn

# Token NGROK
!ngrok authtoken 'INPUT_YOU_OWN_AUTHTOKEN_FROM_NGROK'

# Get file clients.xlsx from my drive
!wget -O clients.xlsx --no-check-certificate -r 'https://drive.google.com/uc?export=download&id=1uDp5OyiMVCBPhaAM0UPI9mIwC70ssmnq'

# Dependences
from fastapi import FastAPI
import nest_asyncio
from pyngrok import ngrok
import uvicorn
import pandas as pd

app = FastAPI()

dir_sheet = "/content/clients.xlsx"
sheet_db = pd.read_excel(dir_sheet)
clients = []


# Create client class
class Client:
    def __init__(self, id, name, age, city, country):
        self.id = id
        self.name = name
        self.age = age
        self.city = city
        self.country = country


# Function to include new client
def new_client():

    for n in sheet_db.index:
        # Get data from sheet
        var_id = sheet_db.iloc[n:n+1, :1].to_string(index=False,
                                                    header=None)
        var_name = sheet_db.iloc[n:n+1, 1:2].to_string(index=False,
                                                       header=None)
        var_age = sheet_db.iloc[n:n+1, 2:3].to_string(index=False,
                                                      header=None)
        var_city = sheet_db.iloc[n:n+1, 3:4].to_string(index=False,
                                                       header=None)
        var_country = sheet_db.iloc[n:n+1, 4:5].to_string(index=False,
                                                          header=None)

        # Use Client class to set variables
        cliente = Client(id=var_id,
                         name=var_name,
                         age=var_age,
                         city=var_city,
                         country=var_country)

        # Include new client into clients list
        clients.append(cliente)


@app.get('/')
async def getj():
    # Clear data from clients list
    clients.clear()
    # Put clients from Sheets into clients clist
    new_client()

    return clients

# Start ngrok
ngrok_tunnel = ngrok.connect(8000)
print('Public URL:', ngrok_tunnel.public_url)
nest_asyncio.apply()
uvicorn.run(app, port=8000)
```
